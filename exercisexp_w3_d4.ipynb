#üåü Exercise 2: Creating a Line Plot for Temperature Variation


Objective: Create a simple line plot using Matplotlib that represents temperature variations over a week.
Tasks:
Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).
Label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.
Add a title to the plot.
Display the plot.


```python
import matplotlib.pyplot as plt


x = [ "Sunday","Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday",]
y = [72, 74, 76, 80, 82, 78, 75]


plt.plot(x, y, '-')


# Label the axes and add a title
plt.xlabel("Day")
plt.ylabel("Temperature (¬∞F)")
plt.title("Temperature Variations Over a Week")

# Display the plot
plt.show()
```

#üåü Exercise 3: Visualizing Monthly Sales with a Bar Chart


Objective: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.
Tasks:

Label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.

```python
import matplotlib.pyplot as plt


months = ['January', 'February', 'March', 'April', 'May']
sales = [5000, 5500, 6200, 7000, 7500]


plt.bar(months, sales, color='pink')


plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.show()
```

#üåü Exercise 4: Visualizing the Distribution of CGPA


Objective: Create a histogram to visualize the distribution of students‚Äô CGPA.

Import necessary libraries.
Use Seaborn‚Äôs histplot to create a histogram of the CGPA categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.DataFrame({'CGPA': [2.75, 3.5,3.0, 3.2,  3.7, 3.8, 5.0, 2.2, 3.9, 3.3, 2.6, 3.1, 3.1, 3.4]})


sns.histplot(df['CGPA'], bins=20, color='purple')


plt.title('distribution of Students')
plt.xlabel('CGPA')
plt.ylabel('count')


plt.show()
```

#üåü Exercise 5: Comparing Anxiety Levels Across Different Genders


Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.
Dataset Overview: The dataset includes columns: ‚ÄòDo you have Anxiety?‚Äô and ‚ÄòChoose your gender‚Äô.
Tasks:
Import necessary libraries.
Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
Customize the plot with an appropriate color palette and add a title.
Display the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import random

data = {
    'Choose your gender': [random.choice(['Male', 'Female']) for _ in range(500)],
    'Do you have Anxiety?': [random.choice(['Yes', 'No']) for _ in range(500)]
}
print(data)
```

```python

df = pd.DataFrame(data)


plt.figure(figsize=(15, 7))
sns.countplot(data=df, x='Choose your gender', hue='Do you have Anxiety?', palette='coolwarm')


plt.title('Comparison of anxiety levels across genders')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Anxiety')


plt.show()
```

#üåü Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from faker import Faker
import random


fake = Faker()


data = {
    'Age': [fake.random_int(min=18, max=60) for _ in range(500)],
    'Do you have Panic Attacks?': [random.choice(['Yes', 'No']) for _ in range(500)]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
import numpy as np
df['Do you have Panic Attacks?'] = np.where(df['Do you have Panic Attacks?'] == 'Yes', 1, 0)
print(df.head(60))
```

```python
# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Age'], y=df['Do you have Panic Attacks?'], alpha=0.5, edgecolor='k')

# Customize plot
plt.xlabel('Age')
plt.ylabel('Panic Attack (Yes=1, No=0)')
plt.title('Age vs Panic Attack Responses')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```
