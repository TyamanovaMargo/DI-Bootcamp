```python
image_path = "/content/flowers/flowers/00_001.png"

original_image = Image.open(image_path)


plt.imshow(original_image)

```

```python
from scipy.ndimage import rotate
import numpy as np

def rotate_image_90_degrees(image):
    image_array = np.array(image)
    rotated_array = rotate(image_array, angle=90, reshape=True, mode='nearest')
    return Image.fromarray(rotated_array)
```

```python
rotated_image = rotate_image_90_degrees(original_image)


plt.imshow(rotated_image)
```
