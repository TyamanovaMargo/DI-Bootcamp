```python
import pandas as pd
train_data = pd.read_csv('2019.csv')
print(train_data.head())
```

```python
#Data Preprocessing


print(train_data.isnull().sum())

#all columns have a count of 0 missing values, meaning there are no NaN (missing) values in the dataset.
```

```python
print(train_data.dtypes)
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.figure(figsize=(8, 5))
plt.scatter(train_data['Social support'], train_data['Score'], color='blue', alpha=0.5)


plt.xlabel("Social Support")
plt.ylabel("Happiness Score")


plt.show()
```

```python
countries = train_data['Country or region'].unique()
num_subplots = len(countries)
print(num_subplots)
print(countries)
```

```python
import math
import matplotlib.pyplot as plt

num_cols = 3
num_rows = math.ceil(num_subplots / num_cols)

plt.figure(figsize=(12, 8))
```

```python
#import matplotlib.pyplot as plt
# mport matplotlib.pyplot as plt
# import numpy as np

# #plot 1:
# x = np.array([0, 1, 2, 3])
# y = np.array([3, 8, 1, 10])

# plt.subplot(1, 2, 1)
# plt.plot(x,y)

# #plot 2:
# x = np.array([0, 1, 2, 3])
# y = np.array([10, 20, 30, 40])

# plt.subplot(1, 2, 2)
# plt.plot(x,y)

# plt.show()

country_data = train_data.groupby('Country or region')[['GDP per capita', 'Healthy life expectancy']].mean()
country_data = country_data.sort_values(by='GDP per capita', ascending=False)


fig, ax = plt.subplots(2, 1, figsize=(30, 10))  # Two subplots: one for GDP, one for Life Expectancy

# First subplot
ax[0].bar(country_data.index, country_data['GDP per capita'], color='pink')
ax[0].set_title('GDP per Capita by Country')
ax[0].set_ylabel('GDP per Capita')
ax[0].tick_params(axis='x', rotation=90)  # Rotate x-axis labels

# Second subplot
ax[1].plot(country_data.index, country_data['Healthy life expectancy'], marker='o', color='purple')
ax[1].set_title('Healthy Life Expectancy by Country')
ax[1].set_ylabel('Healthy Life Expectancy')


plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(40, 8))
ax1.bar(country_data.index, country_data['GDP per capita'], color='pink', label='GDP per Capita')


ax2 = ax1.twinx()
ax2.plot(country_data.index, country_data['Healthy life expectancy'], marker='o', color='purple', label='Healthy life expectancy')

# Labels and title
ax1.set_xlabel('Region')
ax1.set_ylabel('GDP per Capita', color='pink')
ax2.set_ylabel('Healthy life expectancy', color='purple')


# Rotate x-axis labels for better readability
# ax1.set_xticklabels(country_data.index, rotation=90)

# Add grid and legend
ax1.grid(True, linestyle='-', alpha=0.5)
# for more readble x
ax1.set_xticks(range(len(country_data.index)))  # Set tick positions
ax1.set_xticklabels(country_data.index, rotation=45, ha="right")  # Rotate 45° and align right

# Show plot
plt.tight_layout()
plt.show()
```
