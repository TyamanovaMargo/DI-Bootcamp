#Data Loading and Exploration

```python
import pandas as pd

# Load the dataset
file_path = "/content/train(1).csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Info:")
df.info()
print("\nFirst 5 rows:")
print(df.head())

# Display summary statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

#Data Cleaning and Preprocessing:

```python

# Handle missing values
df.fillna(df.median(), inplace=True)  # Replace missing values with median

# Convert categorical data to numerical (if any)
for col in df.select_dtypes(include=['object']).columns:
    df[col] = pd.factorize(df[col])[0]

# Display basic information about the dataset
print("Dataset Info:")
df.info()
print("\nFirst 5 rows:")
print(df.head())

# Display summary statistics
print("\nDescriptive Statistics:")
print(df.describe())
```

#Statistical Analysis with NumPy and SciPy:

```python
import pandas as pd
import scipy.stats as stats
import numpy as np


# Detailed Statistical Analysis
print("\nDetailed Statistical Analysis:")
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    print(f"Feature: {col}")
    print(f"Mean: {df[col].mean()}")
    print(f"Median: {df[col].median()}")
    print(f"Mode: {df[col].mode()[0]}")
    print(f"Range: {df[col].max() - df[col].min()}")
    print(f"Variance: {df[col].var()}")
    print(f"Standard Deviation: {df[col].std()}")
    print(f"Skewness: {df[col].skew()}")
    print(f"Kurtosis: {df[col].kurtosis()}")
    print("-")

# Hypothesis Testing
price_groups = [df[df['price_range'] == i] for i in df['price_range'].unique()]
print("\nANOVA Test for Price Range Groups:")
anova_results = stats.f_oneway(*[group['ram'] for group in price_groups])
print(f"RAM vs Price Range ANOVA p-value: {anova_results.pvalue}")

#  Correlation Analysis
print("\nFeature-Target Correlations:")
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    if col != 'price_range':
        corr, p_value = stats.pearsonr(df[col], df['price_range'])
        print(f"{col}: Pearson Correlation = {corr}, p-value = {p_value}")

# Advanced SciPy
print("\nAdvanced Statistical Insights:")
print("Shapiro-Wilk Test for Normality on RAM:")
shapiro_test = stats.shapiro(df['ram'])
print(f"Shapiro-Wilk p-value: {shapiro_test.pvalue}")
```

#Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Visualization
plt.figure(figsize=(10, 6))
sns.histplot(df['ram'], bins=30, kde=True)
plt.title('RAM Distribution')
plt.xlabel('RAM')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['ram'], y=df['price_range'])
plt.title('RAM vs Price Range')
plt.xlabel('RAM')
plt.ylabel('Price Range')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price_range'], y=df['ram'])
plt.title('Boxplot of RAM across Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('RAM')
plt.show()

plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```
