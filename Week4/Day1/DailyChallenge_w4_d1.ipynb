```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from faker import Faker
```

#1. Data Preparation

```python
fake = Faker()
cities = [fake.city() for _ in range(10)]


months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature_data = np.random.uniform(low=-5, high=35, size=(10, 12))


df = pd.DataFrame(temperature_data, index=cities, columns=months).round(2)




print(df)
```

# Data Analysis

```python
avr_data = df.mean(axis = 1) # average temperature for each city (average by row)
print(avr_data)
```

```python
highest_avg_data = df.idxmax(axis=1)
print(highest_avg_data)

```

```python
lowest_avg_data = df.idxmin(axis=1)
print(lowest_avg_data)
```

```python


highest_in__city = df.mean(axis=1).idxmax()
lowest_in_city = df.mean(axis=1).idxmin()


print(f"Max: {highest_in__city}")
print(f"Min: {lowest_in_city}")
```

```python

avr_data = df.mean(axis=1)

# Город с самой высокой средней температурой
hot_city = avr_data.idxmax()
highest_avg__temp = avr_data.max()

# Город с самой низкой средней температурой
cold_city = avr_data.idxmin()
lowest_avg_temp = avr_data.min()

# Выводим результат
print(f"Hottest city: {hot_city} ({highest_avg__temp:.2f}°C)")
print(f"Coldest city: {cold_city} ({lowest_avg_temp:.2f}°C)")
```

#Data Visualization:

```python
plt.plot(df)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.legend(df.columns, title="City")
plt.title("Monthly Temperature Trends for Each City")
plt.legend(df.index)
```
